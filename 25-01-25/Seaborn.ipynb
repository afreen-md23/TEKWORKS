```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
sns.barplot(data=df, x='Pclass' , y='Survived' , errorbar =None,palette='Blues')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#TASK 
#Is there a relation ship b/w passanger age and fare
sns.scatterplot(data=df, x='Age' , y='Fare',hue= "Survived"  ,palette="deep",alpha=1)
plt.title('relation between passanger age and fares')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
#TASK 
#what are the pairwise relationships b/w age,fare and passenger class ?

sns.pairplot(data=df[['Age', 'Fare', 'Pclass','Sex']], hue='Sex', palette='bright')
plt.show()
```

```python
#TASK
numeric_cols = df.select_dtypes(include='number')

corr_matrix =numeric_cols.corr()
sns.heatmap(corr_matrix,annot=True,cmap='Set2',fmt=".2f",linewidth=0.5)
plt.title("correlation Heatmap")
plt.show()
```
