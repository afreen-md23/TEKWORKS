```python
#here we are importing the matplotlib as plt by using the below command
import matplotlib.pyplot as plt
```

```python
#we have 2 variate analysis
# 1] Univariate:Analysis of Single variable 
# -> ex : Histogram 
#         Box plot
#         Bar chart
#  2] Bivariate:Analysis of Two variables to explore their relationship
#  -> ex: scatter plot : visualizes the relation ship b/w two continuous variables
#         Line Plot: tracks changes in two related variables over time
#         Heat Map : shows cor-realation in a matrix formate
```

```python
# LINE PLOT
# using LinePolt we are exploring the relationship b/w the variables 
#It is a defalut plot 
x=[1,2,3,4,5] 
y=[2,4,6,8,10] 
#creating line plot
#Properites we can use
plt.plot(x,y,color='r', linestyle='--' ,linewidth=2, marker="*", markerfacecolor='green', markersize=8, markeredgecolor='black') 

plt.grid(True)


#Adding title and labels..
plt.title("simple line plot") 
plt.xlabel("x-axis") 
plt.ylabel("y-axis")

# show the plot 

plt.show()
```

```python
#create a line plot showing the growth of a saving account over 6 months with the following 
# x-axis:Month (1-6)
# Y-axis:saving amount

x=[1,2,3,4,5,6]
y=[400,100,1500,600,3000,250]

plt.plot(x,y,color='r', linestyle='--' ,linewidth=2, marker="*",markerfacecolor='green',markersize=8,markeredgecolor='black') 

plt.grid(True)


#Adding title and labels..
plt.title("Growth of a saving account over 6 months") 
plt.xlabel("Months") 
plt.ylabel("Saving Amount")

# show the plot 

plt.show()
```

```python
#BAR CHART 
#data
categories=['A','B','C','D','E']
values=[10,15,7,12,9] 
#create the bar chart 
plt.bar(categories , values ,color=["skyblue","red","green","purple","orange"],edgecolor="black" ,linewidth=5 ) 
#plt.grid(True)

#Adding title and labels 

plt.title("Basic Bar Chart") 
plt.xlabel("Categories") 
plt.ylabel("values")

# show the plot 

plt.show()
```

```python
#HISTOGRAM
#data 

ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,36,38,40,42,44,46,48,49,50,52,53,56,58,60,62,61,64,67,68,69,70,71,72,73,734,75]
b=[20,30,40,50,60,70]

#plot histogram 

plt.hist(ages,b,color="green",edgecolor="black" ,linewidth=2 )
plt.grid(True)
#Adding title and labels 

plt.title("Age Distribution of Customer") 
plt.xlabel("Age ranges") 
plt.ylabel("Frequency")

# show the plot 

plt.show()


```

```python
#PIE CHART 
#data 
programes=["python","java","mysql","Node.js","C++"] 
percent=[60,50,30,20,10]
colors=["skyblue","red","green","purple","orange"]

plt.pie(percent, labels=programes ,colors=colors, startangle=140)

#Adding title  and legend

plt.title("Programming language usages") 

plt.legend()
# show the plot 

plt.show()

```

```python
#SCATTER PLOT

x=[1,2,3,4,5] 
y=[2,6,13,25,10] 
#creating line plot
#Properites we can use
plt.scatter(x,y,color="orange",edgecolor="black" ,linewidth=0.5 ) 

#Adding title and labels..
plt.title("Scatter Plot Example") 
plt.xlabel("x-axis") 
plt.ylabel("y-axis")

# show the plot 

plt.show()
```
